```python
loc='/home/noah/Desktop/large_analysis/ca_analysis/cur_data/'
data=fr.gen_output(loc+'Run_131-all.dat')[0]
data=pd.precuts(data)
data=data[pd.doubles(data)]
data=data.view(recarray)
data.energy=pd.vec_calibrate(data.energy,data.board,data.channel)
```

```python
fs=35
figure(figsize=(30,20))
trutharray=(sim.detector=='E')*(sim.pixel==64)
s,b=histogram(sim.energy,bins=1000,range=[0,800])
d,b=histogram(data.energy,bins=1000,range=[0,800])
b=pd.cbins(b)
s,d=s.astype(float),d.astype(float)
s=s/amax(s)*amax(d[b>50])
plot(b,s,ds='steps')
plot(b,d,ds='steps')
tick_params(labelsize=fs)
yscale('log')
xlim(300,455)
show()
```
