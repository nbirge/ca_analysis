```python
from numpy import *; from numpy.core.records import fromarrays
from matplotlib.pyplot import *
import sys; sys.path.insert(0,'/home/noah/Desktop/large_analysis/ca_analysis/'); import os
import predefined as pd; import fileread as fr
import pickle
import time
```

```python

runs=[40,41,42,43,44,45,46]
data=[]
for run in runs:
    print(run)
    temp=(fr.gen_output('Run_{:0d}-all.dat'.format(run))[0]).view(recarray)
    temp.energy=pd.vec_calibrate(temp.energy,temp.board,temp.channel)
    temp=pd.precuts(temp,energy=20)
    data.append(temp)
data=np.hstack(data)
data=data.view(np.recarray)
```

```python
fs=30
bdchs=[11,12,35]
fig,ax=subplots(nrows=1,figsize=(15,10))
data_sum=zeros(500)
for i,bdch in enumerate(bdchs):
    bd,ch=int(bdch/8),int(bdch%8)
    t=(data.board==bd)*(data.channel==ch)
    data_hist,bins= histogram(data.energy[t],bins=500,range=[0,400])
    if bdch==11:
        data_hist=data_hist.astype(float)
        data_hist*=1.
    data_sum+=data_hist
    bins=pd.cbins(bins)
    ax.plot(bins,data_hist,ds='steps',label=pd.pixel(bd,ch))
    ax.set_title(pd.pixel(bd,ch),fontsize=fs)
    ax.tick_params(labelsize=fs)
    ax.set_xlabel('Energy [keV]',fontsize=fs)
ax.plot(bins,data_sum,ds='steps',label='Sum')
ax.legend(fontsize=fs)
ax.set_xlim(0,260)
ax.set_ylim(1,40000)
show()
```

```python
nbins,rang=1000,[0,5000]
fs=30
fol='./'
runs=[run for run in os.listdir(fol) if run.endswith('-all.dat')]
runs.sort()
bdchs=np.arange(48)
data=np.zeros((48,1000),dtype=float)
for run in runs:
    temp=(fr.gen_output(fol+run)[0]).view(np.recarray)
    for bdch in bdchs:
        bd,ch=divmod(bdch,8)
        if bdch%8==0 or pd.pixel(bd,ch)=='N/A' :
            continue
        t=(temp.board==bd)*(temp.channel==ch)
        data[bdch]+=np.histogram(temp.energy[t],bins=nbins,range=rang)[0]
bins=np.histogram([],bins=nbins,range=rang)[1]
bins=pd.cbins(bins)
```

## West detector 45Ca pixel content

```python
xfac=6.2
bins=np.histogram([],bins=nbins,range=rang)[1]
bins=pd.cbins(bins)
fig,ax=subplots(figsize=(20,10))
for bdch in bdchs:
    bd,ch=divmod(bdch,8)
    if bdch%8==0 or pd.pixel(bd,ch)=='N/A' or bdch>24 :
        continue
    ax.plot(bins/xfac,data[bdch],ds='steps',label=pd.pixel(bd,ch))
# ax.set_yscale('log')
ax.set_title('45Ca runs',fontsize=fs)
ax.legend(fontsize=fs-20,loc='upper right')
ax.set_yscale('log')
ax.set_xlim(0,2e3/xfac)
if xfac<6:
    ax.set_xlabel('ADC',fontsize=fs)
else:
    ax.set_xlabel('~ keV',fontsize=fs)
ax.set_ylim(1,10e5)
ax.tick_params(labelsize=fs)
show()
```

```python
bins=np.histogram([],bins=nbins,range=rang)[1]
bins=pd.cbins(bins)
fig,ax=subplots(figsize=(20,10))
for bdch in bdchs:
    bd,ch=divmod(bdch,8)
    if bdch%8==0 or pd.pixel(bd,ch)=='N/A' :
        continue
    if pd.pixel(bd,ch) in ['75W','52W','39W','50W','53W','78W','41W']:
        ax.plot(bins/xfac,data[bdch],ds='steps',label=pd.pixel(bd,ch))
# ax.set_yscale('log')
ax.set_title('45Ca runs',fontsize=fs)
ax.legend(fontsize=fs-20,loc='upper right')
ax.set_yscale('log')
ax.set_xlim(0,2e3/xfac)
if xfac<6:
    ax.set_xlabel('ADC',fontsize=fs)
else:
    ax.set_xlabel('~ keV',fontsize=fs)
ax.set_ylim(1,1e5)
ax.tick_params(labelsize=fs)
show()
```
