```python
from numpy import *; from matplotlib.pyplot import *
import sys; sys.path.insert(0,'/home/noah/Desktop/large_analysis/ca_analysis/')
import predefined as pd; import fileread as fr; import wave_ops as wo

path='/media/noah/Data/calcium_WF_data/'

def grab_run(run=120,numwaves=10000):
    data=fr.raw(path+'Run_{:0d}_0.bin'.format(run),numwaves=numwaves)
    wo.baseline_restore(data)
    return data.view(recarray)

def grab_pixel(data,bd,ch):
    t=(data['board']==bd)*(data['channel']==ch)
    return data[t]
    
```

```python
x=grab_run(numwaves=100000)
y=grab_pixel(x,1,2)

```

```python
t=(np.amax(y.wave,axis=1)>200)*(np.amax(y.wave,axis=1)<300)
fig,ax=subplots(figsize=(20,10))
ax.plot(y.wave[t][1])

show()
```
