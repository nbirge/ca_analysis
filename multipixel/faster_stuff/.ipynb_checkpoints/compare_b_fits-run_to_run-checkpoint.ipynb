```python
from numpy import *; from matplotlib.pyplot import *

labels=['Data','b=0','b=0.4','b=1']

bdists=[]
bdists.append(load('b_s-faster.npy'))
bdists.append(load('b_s-b_0.0-s_10000.npy'))
bdists.append(load('b_s-b_0.4-s_10000.npy'))
# bdists.append(load('b_s-b_0.7-s_10000.npy'))
bdists.append(load('b_s-b_1.0-s_10000.npy'))

means=[]
stds=[]
for dist in bdists:
    means.append(np.mean(dist['b']))
    stds.append(np.std(dist['b']))
    
```

```python
nbins=201;rang=[-1,1]
fig,ax=subplots(figsize=(15,10))
for i,dist in enumerate(bdists):
    ax.hist(dist['b'],bins=nbins,range=rang,histtype='step',\
            label=labels[i]+'\n'+r'$\mu$={:0.2f} $\sigma=${:0.2e}'.format(means[i],stds[i]),density=True)
ax.set_xlim(0,1)
ax.legend()
ax.set_xlabel('b')
ax.set_ylabel('Arb. Units')
show()
```
