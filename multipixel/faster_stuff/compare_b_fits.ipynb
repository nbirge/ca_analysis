```python
from numpy import *; from matplotlib.pyplot import *

labels=arange(10)

bdists=[]
for label in labels:
    bdists.append(load('b_s-b_0.4-s_1000-r_{:d}.npy'.format(label)))
means=[]
stds=[]
for dist in bdists:
    means.append(np.mean(dist['b']))
    stds.append(np.std(dist['b']))
    
```

```python
nbins=101;rang=[-1,1]
fig,ax=subplots(figsize=(15,10))
for i,dist in enumerate(bdists):
    ax.hist(dist['b'],bins=nbins,range=rang,histtype='step',\
            label='Run {:d}'.format(i)+'\n'+r'$\mu$={:0.2f} $\sigma=${:0.2e}'.format(means[i],stds[i]),density=True)
ax.set_xlim(0,1)
ax.legend()
ax.set_xlabel('b')
ax.set_ylabel('Arb. Units')
show()
```
