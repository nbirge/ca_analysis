```python
from numpy import *; from matplotlib.pyplot import *

fs=30
threshes=arange(30,100,10)

bs={}
for thresh in threshes:
    bs.update({thresh:np.load('b_vals-T_'+str(thresh)+'.npy').view(recarray)})

fig,ax=subplots(figsize=(30,10))

for thresh in threshes:
    ax.errorbar(bs[thresh].beg,bs[thresh].b,bs[thresh].b_error,fmt='o',label='Threshold: {:0d} [keV]'.format(thresh))
ax.tick_params(labelsize=fs)
ax.set_ylim(-1,2)
ax.set_xlim(55,155)
ax.legend(fontsize=fs)
# ax.axhline(0,0,200,ls='--',c='0.5')
ax.grid()
ax.set_xlabel('Beginning of Fit [keV]\n Note: Fit range 100 keV wide',fontsize=fs)
ax.set_ylabel('Extracted \'b\' value',fontsize=fs)
fig.tight_layout()
savefig('b_thresh_comp')
show()
```
